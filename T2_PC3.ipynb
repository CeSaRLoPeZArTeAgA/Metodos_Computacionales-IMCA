#1.-Completar el codigo para hacer el retroceso cuando ya no hay mas opciones en la fila.

```python
# FunciÃ³n para verificar si es seguro colocar una reina en una posiciÃ³n dada
def es_seguro(tablero, fila, columna, N):
    # Verificar columna hacia arriba
    for i in range(fila):
        if tablero[i][columna] == 1:
            return False

    # Verificar diagonal superior izquierda
    i, j = fila, columna
    while i >= 0 and j >= 0:
        if tablero[i][j] == 1:
            return False
        i -= 1
        j -= 1

    # Verificar diagonal superior derecha
    i, j = fila, columna
    while i >= 0 and j < N:
        if tablero[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True

# FunciÃ³n principal para resolver el problema de las N reinas y encontrar todas las soluciones
def resolver_n_reinas_todas(tablero, fila, N, soluciones):
    # Si todas las reinas estÃ¡n colocadas, almacena la soluciÃ³n
    if fila == N:
        soluciones.append([row[:] for row in tablero])
        return

    # Intenta colocar una reina en cada columna de la fila actual
    for columna in range(N):
        if es_seguro(tablero, fila, columna, N):
            # Coloca la reina
            tablero[fila][columna] = 1

            # Llama recursivamente para colocar reinas en las siguientes filas
            resolver_n_reinas_todas(tablero, fila + 1, N, soluciones)

            # Retrocede (backtracking)
            tablero[fila][columna] = 0
```

#2.-Realice el diagrama anterior para 5 reinas y ejecute el codigo paso a paso. Determine todas las soluciones.

```python
# FunciÃ³n para imprimir una soluciÃ³n de las reinas
def imprimir_soluciones(soluciones):
    for index, solucion in enumerate(soluciones):
        print(f"SoluciÃ³n {index + 1}:")
        for fila in solucion:
            print(" ".join("R" if x == 1 else "." for x in fila))
        print("\n")

# Resolver para el caso de 5 reinas
N = 5
tablero = [[0 for _ in range(N)] for _ in range(N)]
soluciones = []

# Encontrar todas las soluciones
resolver_n_reinas_todas(tablero, 0, N, soluciones)

# NÃºmero total de soluciones
if len(soluciones)!=0:
  print("SOLUCIONES TOTALES")
  # Imprimir todas las soluciones
  imprimir_soluciones(soluciones)

print("Numero total de Soluciones:",len(soluciones))
```
